# *Data Science in Medical-Live Tracker*

This analysis deals with the analysis of the Covid-19 cases in India 

```
import pandas as pd

data = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
data
```

# *Data Contains*

* state - provience in the India
* active - the number of active covid cases
* cured - the number of the cured covid cases in India
* death - the number of people who died due to the Covid-19
* new_active - the number of the cases which are added 
* new_positive - the number of new covid positive cases within 24 hours
* new_cured - the number of new cured cases within 24 hours
* new_death - the number of new death cases within 24 hours 
* state_code - the code which is provided by the India Ministry
